```python
# Q3) Which genre had highest number of releases


df1['Genre'].value_counts().idxmax()
```

```python
# Q4) How many movies in each genre got released in different release times like long 
#     weekend, festive season, etc. (Note: Do a cross tabulation between Genre and ReleaseTime.)


pd.crosstab(df1.Genre,df1.ReleaseTime)
```

```python
#to convert release date into datetime data type
df1['Release Date']=pd.to_datetime(df1['Release Date'])
df1
```

```python
# Q5) Which month of the year, maximum number movie releases are seen? (Note: Extract a new column called month from ReleaseDate column.)
# Exracting new column

df1['Month']=df1['Release Date'].dt.month
df1
```

```python
df1['Month'].value_counts().idxmax()
```

```python
# Q6) Which month of the year typically sees most releases of high budgeted movies, that is,movies with budget of 25 crore or more?

df1['Month'].loc[df1['Budget']>25]
```

```python
# Q7) Which are the top 10 movies with maximum return on investment (ROI)? Calculate return on investment (ROI) as (BoxOfficeCollection – Budget) / Budget.

df1['roi']=(df1['BoxOfficeCollection']-df1['Budget'])/df1['Budget']
```

```python
x=df1.sort_values(['roi'],ascending=False).head(10)
x['MovieName']
```

```python
# Q8) Do the movies have higher ROI if they get released on festive seasons or long weekend? Calculate the average ROI for different release times

pd.crosstab(df1.roi.max(),df1.ReleaseTime)

```

```python
df1.groupby(['ReleaseTime']).roi.mean()
```

```python
# Q9) Is there a correlation between box office collection and YouTube likes? Is the correlation positive or negative?

x = df1['BoxOfficeCollection'].corr(df1['YoutubeLikes'])
if (x>=0):
 (print('Correlation is ' ,x,' and it is  positive'))
else:
 (print('correlation is ' ,x ,' and it is negative'))
```
