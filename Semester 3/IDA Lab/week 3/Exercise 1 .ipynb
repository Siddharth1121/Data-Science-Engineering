```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sms
```

```python
# importing mtcars.csv file


df=pd.read_csv("mtcars.csv")
df
```

```python
# Q1) Find the car with the best mpg

bestmpg=df['model'].loc[df['mpg'].idxmax()]
print('car with the best mpg is :',bestmpg)
```

```python
# Q2) Find the car with the worst mpg

worstmpg=df['model'].loc[df['mpg'].idxmin()]
print('car with the worst mpg is :',worstmpg)
```

```python
# Q3) Find the car with the best horsepower

besthp=df['model'].loc[df['hp'].idxmax()]
print('car with the best horsepower is :',besthp)
```

```python
# Q4) Find 5 number summary of displacement

summary=df.describe()
summary
```

```python
# Q5) Find median horse power

medianhp=df['hp'].median()
print('Median horse power is :',medianhp)
```

```python
# Q6) What is average mpg for manual vs. automatic cars

avg=df.loc[df['am']==0]
manualavg=avg.mpg.mean()
print('The average mpg for manual cars is :')
manualavg
```

```python
# Q6) What is average mpg for manual vs. automatic cars

avg=df.loc[df['am']==1]
autoavg=avg.mpg.mean()
print('The average mpg for automatic cars is :')
autoavg
```

```python
# Q7) Draw a histogram of miles per gallon

histompg=df['mpg'].hist()
histompg
plt.title('Histogram of miles per gallon')
plt.xlabel('miles')
plt.ylabel('gallons')
```

```python
# Q8)  Boxplot of mpg for each cylinder type

boxplotmpg=sms.boxplot(x="cyl",y="mpg",data=df).set(title="Boxplot of mpg for each cylinder type")
boxplotmpg
```

```python
# Q9) Create a crosstab displaying count of automatic vs. manual cars

x=pd.crosstab(df.am, df.am)
x
```

```python
# Q10) Create a crosstab displaying count of â€œam vs cylâ€

x=pd.crosstab(df.am,df.cyl)
x
```

```python
# Q11) What is the correlation between the weight of the car and mpg

df[['wt','mpg']].corr()
```
